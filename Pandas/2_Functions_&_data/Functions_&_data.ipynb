```python
import pandas as pd
raiting=pd.read_csv('ratings.csv')
raiting.head()
```

```python
def class_movies(row):
    if row['rating']<=2.0:
        return 'низкий рейтинг'
    elif 2.0<row['rating']<=4.0:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
raiting['class'] = raiting.apply(class_movies, axis=1)
raiting.head()
```

```python
import pandas as pd
geo_data = {

'Центр':['москва', 'тула', 'ярославль'],

'Северо-Запад':['петербург', 'псков', 'мурманск'],

'Дальний Восток':['владивосток', 'сахалин', 'хабаровск'] 
}
towns=[]
for values in geo_data.values():
    for i in values:
        towns.append(i)
towns
```

```python
row='вконтакте моя страница москва'
data=row.split(' ')
def check(t):
    for i in t:
        if i in towns:
            for items in geo_data.items():
                if i in items[1]:
                    print(items[0])
check(data)
```

```python
df=pd.read_csv('keywords.csv')
df.head()
```

```python
def region_in_keyword(row):
    data=row['keyword'].split(' ')
    i=0
    for word in data:
        if word in towns:
            for items in geo_data.items():
                if word in items[1]:
                    i+=1
                    return items[0]
    if i==0:
        return 'undefined'       
```

```python
df['region'] = df.apply(region_in_keyword, axis=1)
df.head()
```

```python
import pandas as pd
movies_dt = pd.read_csv('movies.csv')
ratings_dt = pd.read_csv('ratings.csv')

# создаем общую таблицу Х
x = ratings_dt.merge(movies_dt, how='left')

# делаем таблицу индекс- название фильма - средний ретинг
films_rating = x.groupby('title').mean().reset_index()[['title', 'rating']]

films_rating.head()
```

```python
""" 1. В переменную years запишите список из всех годов с 1950 по 2010"""
years = [str(i+1950) for i in range(61)]
print(years)
```

```python
""" 2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска.
    Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
     - для каждой строки пройдите по всем годам списка years
     - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
     - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

 создаем функцию с вариантами оценки """

def production_year(row):
    for year in years:
        if year in row['title']:
            return year
    return '1900'
```

```python
""" 3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year' """
films_rating['year'] = films_rating.apply(production_year, axis=1)
films_rating.head(10)
```
