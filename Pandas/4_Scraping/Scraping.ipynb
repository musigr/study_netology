```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://habr.com/ru/search/'
params = {'q': "list",
          'target_type': 'posts',
          'order': 'relevance'}
          
res = requests.get(URL, params)
```

```python
soup = BeautifulSoup(res.text)
link_blocks = soup.find_all("a", class_="tm-article-snippet__title-link")
```

```python
time_blocks = soup.find_all("time")
habr_posts = pd.DataFrame()
for i in range(len(time_blocks)):
    dte = time_blocks[i]['title'][:10]
    lnk = "https://habr.com"+link_blocks[i]['href']
    ttl = str(link_blocks[i].span)[6:][:-7].replace('</em>','').replace('<em class="searched-item">','')
         
    row = {'id': i+1, 'Дата': dte, 'Заголовок': ttl, 'Ссылка на материал': lnk}
    habr_posts = pd.concat([habr_posts, pd.DataFrame([row])])

habr_posts
```
